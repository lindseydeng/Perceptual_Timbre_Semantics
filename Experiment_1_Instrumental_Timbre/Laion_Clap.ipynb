```python
import os
import numpy as np
import pandas as pd
import torch
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import laion_clap

# --------------------------------------------------
# 1. Load LAION-CLAP model
# --------------------------------------------------
model = laion_clap.CLAP_Module(enable_fusion=False)
model.load_ckpt()

# --------------------------------------------------
# 2. Define paths
# --------------------------------------------------
western_csv_path = "/Users/lindseydeng/Desktop/Timbre_Study/timbre_semantics_exp1/instrument_timbre/data/Western.csv"
chinese_csv_path = "/Users/lindseydeng/Desktop/Timbre_Study/timbre_semantics_exp1/instrument_timbre/data/Chinese.csv"

western_audio_folder = "/Users/lindseydeng/Desktop/Timbre_Study/timbre_semantics_exp1/instrument_timbre/data/audio/Western"
chinese_audio_folder = "/Users/lindseydeng/Desktop/Timbre_Study/timbre_semantics_exp1/instrument_timbre/data/audio/Chinese"

output_dir = "/Users/lindseydeng/Desktop/Timbre_Study/timbre_semantics_exp1/results/LAION_CLAP"
os.makedirs(output_dir, exist_ok=True)

# --------------------------------------------------
# 3. Load CSVs and get audio paths
# --------------------------------------------------
df_western = pd.read_csv(western_csv_path)
df_chinese = pd.read_csv(chinese_csv_path)

western_paths = [os.path.join(western_audio_folder, f"{int(row['instrument_id'])}.wav") for _, row in df_western.iterrows()]
chinese_paths = [os.path.join(chinese_audio_folder, f"{int(row['instrument_id'])}.wav") for _, row in df_chinese.iterrows()]

# --------------------------------------------------
# 4. Get text and audio embeddings
# --------------------------------------------------
descriptors = df_western.columns[2:].tolist()

text_embeddings = model.get_text_embedding(descriptors, use_tensor=False)
text_embeddings = np.array(text_embeddings)

western_audio_embeddings = model.get_audio_embedding_from_filelist(x=western_paths, use_tensor=False)
chinese_audio_embeddings = model.get_audio_embedding_from_filelist(x=chinese_paths, use_tensor=False)

# --------------------------------------------------
# 5. Compute cosine similarities
# --------------------------------------------------
def get_similarity_matrix(audio_embeddings, text_embeddings):
    return cosine_similarity(audio_embeddings, text_embeddings)

sim_western = get_similarity_matrix(western_audio_embeddings, text_embeddings)
sim_chinese = get_similarity_matrix(chinese_audio_embeddings, text_embeddings)

similarity_df_western = pd.DataFrame(sim_western, columns=descriptors, index=df_western['instrument_name'])
similarity_df_chinese = pd.DataFrame(sim_chinese, columns=descriptors, index=df_chinese['instrument_name'])

# --------------------------------------------------
# 6. Descriptor-level correlation (across instruments)
# --------------------------------------------------
combined_ratings = pd.concat([df_chinese[descriptors], df_western[descriptors]], ignore_index=True)
combined_similarities = pd.concat([similarity_df_chinese[descriptors], similarity_df_western[descriptors]], ignore_index=True)

desc_level_corrs = []
for desc in descriptors:
    human = combined_ratings[desc].values
    model = combined_similarities[desc].values

    pearson_corr, _ = pearsonr(human, model)
    spearman_corr, _ = spearmanr(human, model)

    desc_level_corrs.append({
        "descriptor": desc,
        "pearson": pearson_corr,
        "spearman": spearman_corr
    })

desc_level_df = pd.DataFrame(desc_level_corrs)
desc_level_df.to_csv(os.path.join(output_dir, "laion_descriptor_correlation_combined.csv"), index=False)

# --------------------------------------------------
# 7. Instrument-level correlation (semantic profiles)
# --------------------------------------------------
def compute_instrument_profile_correlations(human_df, model_df, name_col):
    correlations = []
    for idx, row in human_df.iterrows():
        human_vector = row[descriptors].values
        model_vector = model_df.iloc[idx][descriptors].values

        pearson_corr, _ = pearsonr(human_vector, model_vector)
        spearman_corr, _ = spearmanr(human_vector, model_vector)

        correlations.append({
            "instrument": row[name_col],
            "pearson": pearson_corr,
            "spearman": spearman_corr
        })
    return pd.DataFrame(correlations)

chinese_instr_corr_df = compute_instrument_profile_correlations(df_chinese, similarity_df_chinese, "instrument_name")
western_instr_corr_df = compute_instrument_profile_correlations(df_western, similarity_df_western, "instrument_name")
combined_instr_corr_df = pd.concat([chinese_instr_corr_df, western_instr_corr_df], ignore_index=True)

# Export
chinese_instr_corr_df.to_csv(os.path.join(output_dir, "laion_profile_correlation_chinese.csv"), index=False)
western_instr_corr_df.to_csv(os.path.join(output_dir, "laion_profile_correlation_western.csv"), index=False)
combined_instr_corr_df.to_csv(os.path.join(output_dir, "laion_profile_correlation_combined.csv"), index=False)

# --------------------------------------------------
# 8. Plot Descriptor-Level Correlations
# --------------------------------------------------
plt.figure(figsize=(12, 6))
plt.bar(desc_level_df['descriptor'], desc_level_df['pearson'])
plt.ylabel('Pearson Correlation')
plt.title('LAION-CLAP Similarity vs Human Ratings per Descriptor (All Instruments)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "laion_descriptor_correlation_combined.png"))
plt.show()

# --------------------------------------------------
# 9. Plot Instrument-Level Correlations

plt.figure(figsize=(12, 6))
plt.bar(chinese_instr_corr_df['instrument'], chinese_instr_corr_df['pearson'])
plt.ylabel('Pearson Correlation')
plt.title('LAION-CLAP vs Human Semantic Profile per Instrument (Chinese Instruments)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "laion_profile_correlation_per_instrument_chinese.png"))
plt.show()


plt.figure(figsize=(12, 6))
plt.bar(western_instr_corr_df['instrument'], western_instr_corr_df['pearson'])
plt.ylabel('Pearson Correlation')
plt.title('LAION-CLAP vs Human Semantic Profile per Instrument (Western Instruments)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "laion_profile_correlation_per_instrument_western.png"))
plt.show()
```
